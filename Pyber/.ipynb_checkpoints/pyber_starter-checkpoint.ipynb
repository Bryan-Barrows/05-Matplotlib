```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_data_to_load = "data/city_data.csv"
ride_data_to_load = "data/ride_data.csv"

# Read the City and Ride Data
city_data = pd.read_csv(city_data_to_load)
ride_data = pd.read_csv(ride_data_to_load)


# Combine the data into a single dataset
cityRide_df = city_data.merge(ride_data, on='city', how='left')

# Display the data table for preview
cityRide_df.head()
```

```python
# average fare per city and type
city_group = cityRide_df.groupby(['city','type'])
average_fare = city_group['fare'].mean().reset_index()
avg_fare_type = average_fare.set_index('type')
avg_fare_urban = avg_fare_type.loc['Urban']
avg_fare_rural = avg_fare_type.loc['Rural']
avg_fare_suburban = avg_fare_type.loc['Suburban']

# total no. of rides per city
count_rides = city_group['ride_id'].count().reset_index()
count_rides_type = count_rides.set_index('type')
count_rides_urban = count_rides_type.loc['Urban']
count_rides_rural = count_rides_type.loc['Rural']
count_rides_suburban = count_rides_type.loc['Suburban']

# total no. of drivers per city
drivers = city_data[['city','driver_count','type']]
drivers_type = drivers.set_index('type')
drivers_urban = drivers_type.loc['Urban']
drivers_rural = drivers_type.loc['Rural']
drivers_suburban = drivers_type.loc['Suburban']
```

## Bubble Plot of Ride Sharing Data

```python
# Obtain the x and y coordinates for each of the three city types

# Urban cities
x_axis_urban = count_rides_urban['ride_id']
y_axis_urban = avg_fare_urban['fare']
plt.scatter(x_axis_urban, y_axis_urban, color='lightskyblue',s=drivers_urban['driver_count'], alpha=1.0,label="Urban")

# Rural cities
x_axis_rural = count_rides_rural['ride_id']
y_axis_rural = avg_fare_rural['fare']
plt.scatter(x_axis_rural, y_axis_rural, color='gold',s=drivers_rural['driver_count'], alpha=1.0,label="Rural")

# Suburban cities
x_axis_suburban = count_rides_suburban['ride_id']
y_axis_suburban = avg_fare_suburban['fare']
plt.scatter(x_axis_suburban, y_axis_suburban, color='lightcoral',s=drivers_suburban['driver_count'], alpha=1.0,label="Suburban")

# adding labels & other formatting
plt.legend(loc='best')
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average Fare ($)")
plt.title("Ride Sharing Data")
plt.grid(True)

# Incorporate a text label regarding circle size
plt.text(45, 30, "Note: Circle size correlates with driver count per city")

# Save Figure
plt.show()
```

## Total Fares by City Type

```python
# Calculate Type Percents
total_fares = cityRide_df['fare'].sum()
city_type = cityRide_df.groupby('type')
total_fares_type = city_type.sum().reset_index()

type_index = total_fares_type.set_index('type')
rural_fare = type_index.loc[['Rural'],['fare']]
percent_rural_fare = ((rural_fare['fare'])/(total_fares)*100)
print(percent_rural_fare)

urban_fare = type_index.loc[['Urban'],['fare']]
percent_urban_fare = ((urban_fare['fare'])/(total_fares)*100)
print(percent_urban_fare)

suburban_fare = type_index.loc[['Suburban'],['fare']]
percent_suburban_fare = ((suburban_fare['fare'])/(total_fares)*100)
print(percent_suburban_fare)

# Build Pie Chart
labels = ['Rural','Urban','Suburban']
sizes = [percent_rural_fare, percent_urban_fare, percent_suburban_fare]
colors = ['gold', 'lightskyblue', 'lightcoral']
explode=(0,0.1,0)

plt.pie(sizes, explode=explode, labels=labels, colors=colors, autopct="%1.1f%%", shadow=True, startangle=140)
plt.axis('equal')
plt.title('Percentage of Total Fares by City Type')

# Save Figure
plt.show()
```
